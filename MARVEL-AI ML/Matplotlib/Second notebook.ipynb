```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
pet=pd.read_csv('petrol.csv')

plt.title('Petrol prices over time')

plt.xlabel('Price in rupees per litre')
plt.ylabel('Date')

plt.plot(pet.Delhi,pet.Date, label='Delhi')
plt.plot(pet.Kolkata,pet.Date, label='Kolkata')
plt.plot(pet.Mumbai,pet.Date, label='Mumbai')
plt.plot(pet.Chennai,pet.Date, label='Chennai')

plt.yticks(pet.Date[::50])

plt.legend()

plt.show()
```

```python
fifa=pd.read_csv('fifa_data.csv')
fifa.head(5)
```

```python
bins=[30,40,50,60,70,80,90,100]

plt.hist(fifa.Overall, bins=bins)

plt.xticks(bins)

plt.ylabel('No. of players')
plt.xlabel('Skill level')
plt.title('Distribution of player skills in FIFA 2018')

plt.show()
```

```python
left= fifa.loc[fifa['Preferred Foot']== 'Left'].count()[0]
right=fifa.loc[fifa['Preferred Foot']== 'Right'].count()[0]

labels=['Left', 'Right']

plt.pie([left, right], labels=labels, autopct='%.2f %%')

plt.title('Foot Preference of FIFA players')

plt.show()
```

```python
gas=pd.read_csv('gas_prices.csv')

plt.title('Gas prices over time in USD')

plt.xlabel('Year')
plt.ylabel('Price in USD')

plt.xticks(gas.Year[::3])

plt.scatter(gas.Year,gas.Australia, alpha=0.5, label='Australia')
plt.scatter(gas.Year,gas.Canada, alpha=0.5, label='Canada')
plt.scatter(gas.Year,gas.USA, alpha=0.5, label='USA')
plt.scatter(gas.Year,gas.UK, alpha=0.5, label='UK')

plt.legend()

plt.show()
```
